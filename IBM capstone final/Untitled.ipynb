```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Algorithm': ['KNNBasic', 'NMF', 'ANN', 'Linear Regression', 'Ridge Regression', 'Lasso Regression', 'ElasticNet'],
    'RMSE': [1.01284, 0.2031, 0.0143, 0.210290, 0.210290, 0.210945, 0.210945]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
plt.bar(df['Algorithm'], df['RMSE'])
plt.xlabel('Algorithm')
plt.ylabel('RMSE')
plt.title('RMSE Comparison')
plt.xticks(rotation=45)
plt.show()
```
